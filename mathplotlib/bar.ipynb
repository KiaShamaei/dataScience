```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
x= np.arange(0,10)
y = np.random.uniform(40,60,(10,))
plt.bar(x,y)
```

```python
x= np.arange(0,10)
y = np.random.uniform(40,60,(10,))
#add bottom custom 
plt.bar(x,y, bottom=10)
plt.ylim(0,100)

```

```python
x= np.arange(0,10)
y = np.random.uniform(40,60,(10,))
#add bottom custom list
plt.bar(x,y, bottom=np.random.uniform(10,60,(10,)))
plt.ylim(0,100)
```

```python
x= np.arange(0,10)
y = np.random.uniform(40,60,(10,))
#add bottom custom list and hirzuntal
plt.barh(x,y)
```
