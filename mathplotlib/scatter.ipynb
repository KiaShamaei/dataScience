```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
y = np.random.normal(0,10,(1000,))
x= np.arange(1000)
plt.scatter(x,y)
plt.plot([0,1000] ,[10,10] , '--g')
plt.plot([0,1000] ,[-10,-10] , '--y')
plt.plot([0,1000] ,[0,0] , '--r');
```

```python
y = np.random.uniform(0,10,(1000,))
x= np.arange(1000)
plt.scatter(x,y)
plt.plot([0,1000] ,[10,10] , '--g')
plt.plot([0,1000] ,[-10,-10] , '--y')
plt.plot([0,1000] ,[0,0] , '--r');
```

```python
x = np.linspace(0,360,50)
y = np.sin(np.radians(x))
plt.scatter(x,y , c=['b','r']*25) ;
```

```python
x = np.linspace(0,360,50)
y = np.sin(np.radians(x))
#default color map is viridis and we can change it
plt.scatter(x, y , c=np.arange(0,50) );
```

```python
x = np.linspace(0,360,50)
y = np.sin(np.radians(x))
#default color map is viridis and we can change it and matplotlib normalize it 
plt.scatter(x, y , c=np.arange(0,50) , cmap='hsv');
```

```python
x = np.linspace(0,360,50)
y = np.sin(np.radians(x))
#for make cross color use a trick 
plt.scatter(x, y , c=[10,12]*25 )
```

```python
x = np.linspace(0,360,50)
y = np.sin(np.radians(x))
#default color map is viridis and we can change to prevent normalize default we should add vmax and vmin
plt.scatter(x, y , c=np.arange(0,50) , cmap='plasma' , vmin= 0 , vmax=100)
```
