```python

import matplotlib.pyplot as plt
import numpy as np
```

```python
x= np.linspace(0,360,100)
y= np.sin(np.radians(x))
plt.plot(x,y)
```

```python
plt.plot(x,y,'--')
```

```python
plt.plot(x,y,'.')
```

```python
plt.plot(x,y,'or')
```

```python
# for make vectors in same page just add it in a cell 
y1 = np.cos(np.radians(x))
plt.plot(x,y,'--g')
plt.plot(x,y1,'y')
```
