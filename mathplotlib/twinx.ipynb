```python
import matplotlib.pyplot as plt
import numpy as np
```

make a make to figure in a plot

```python
x = np.linspace(0,360,100)
y = np.sin(np.radians(x))
y1 = np.cos(np.radians(x))
plt.figure(figsize=(15,3) , dpi=72)
plt.subplot(1,2,1)
plt.plot(x,y , 'Red')
plt.subplot(1,2,2)
plt.plot(x,y1, )
```

```python
plt.figure(figsize=(15,3) , dpi=72)
plt.subplot(2,1,1)
plt.plot(x,y , 'Red')
plt.subplot(2,1,2)
plt.plot(x,y1, )
```

```python
from PIL import Image
```

```python
image = np.array(Image.open('1.jpg'))
plt.imshow(image);
```

```python
plt.figure(figsize=(15,9) , dpi=72)
plt.subplot(3,3,2)
plt.imshow(image);
plt.suptitle("use image to show subplot")

plt.subplot(3,3,4)
plt.imshow(image[:,:,0] , 'Reds');
plt.xticks([])
plt.yticks([])
plt.title("red")
plt.subplot(3,3,5)
plt.imshow(image[:,:,1] , 'Blues');
plt.xticks([])
plt.yticks([])
plt.title("blue")
plt.subplot(3,3,6)
plt.imshow(image[:,:,2] , 'Greens')
plt.xticks([])
plt.yticks([])
plt.title("green");
```
