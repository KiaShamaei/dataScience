```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.linspace(0,360,100)
y= np.sin(np.radians(x))
plt.plot(x,y)
```

```python
x = np.linspace(0,360,50)
y= np.sin(np.radians(x))
plt.plot(x,y, '^-g')
```

```python
x = np.linspace(0,360,30)
y= np.sin(np.radians(x))
y1= np.cos(np.radians(x))
plt.plot(x,y, '.-b')
```

```python
plt.plot(x,y, '.-r')
plt.plot(x,y1)
plt.savefig('testSaveFig.svg')
```
