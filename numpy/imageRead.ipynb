### Read image make array of them and show each layers
```
we should use pillow to read data of image and use
matplotlib to show it PIL is package name of pillow 
```

````python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
````

````python
img = Image.open('test.jpg')
#make nd array from it 
im = np.array(img)
im.shape
````

````python

n = im.max(axis=2)
plt.imshow(n,'gray')
````

````python
b= im[:,:,0] + im[:,:,1] + im[:,:,2]
plt.imshow(b , 'gray')
````

````python
# so do it mean 
c= im[:,:,0]//3 + im[:,:,1]//3 + im[:,:,2]//3
plt.imshow(c , 'gray')
````

````python
#best practice 
e= im[:,:,0]*0.2126 + im[:,:,1]*0.7152 + im[:,:,2]*0.07152
plt.imshow(e,'gray')
````
