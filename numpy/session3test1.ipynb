### How to make to np array mapping to find answer

```python
users = np.array(['ali' , 'mahsa','reza'])
rankOfUser = np.array([0 , 1 , -1])
# now find user who rank is more than zero 
users[rankOfUser > 0]
```

```python
#find max rank 
maxRank = np.max(rankOfUser)
filterMax = ( rankOfUser == maxRank)
users[filterMax]
```
