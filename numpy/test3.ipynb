```python
import random
nums= []
for i in range(1000):
    nums.append(random.randint(1,100))
counts= {}
for n in nums:
    if n in counts:
        counts[n] += 1 
    else:
        counts[n] = 1

```

```python
print(counts)
```

```python
import matplotlib.pyplot as pt 
pt.ylim(0,1000)
pt.plot(counts.values())
```
