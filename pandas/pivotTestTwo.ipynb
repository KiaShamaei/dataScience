```python
#we need openpyxl for work xls
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

---

First read excle file after pivot on it and show them in plot

```python
df = pd.read_excel('sales-funnel.xlsx')
```

```python
df
```

```python
nt = pd.pivot_table(df,values=['Price'] , index=['Name'] , aggfunc=np.sum)
```

```python
nt.plot()
```

```python
nt2 = pd.pivot_table(df , index=['Name' , 'Manager' ,'Rep'] , aggfunc=np.sum)
```

```python
nt2
```

```python
nt2.plot()
```

make it more read by chage param in function

```python
nt3 = pd.pivot_table(df , index=[ 'Manager' ,'Rep'] , values=['Price'] , aggfunc=np.sum)
```

```python
nt3
```

```python
# Create a figure with a custom size
plt.figure(figsize=(12, 6))  # width, height in inches

# Plot the DataFrame
nt3.plot()

# Customize the plot
plt.xlabel('Manager')
plt.ylabel('Price')
plt.title('Sample DataFrame Plot')

# Display the plot
plt.show()
nt3.to_csv('test_csv_df.csv')
```
