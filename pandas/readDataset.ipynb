### Map list python 
$ map f(x) $
There are for solution to map a dataSeries :
- use normal loop and if condition
- use list comprehension
- use lamdba on map 

```python
a = ["Y","N","N","yes","na","Y"]
a_prime = []
for i in a :
    if i not in ["Y","N"]:
        a_prime.append("NA")
    else:
        a_prime.append(i)
a_prime
```

```python
#List Comprehenstion
a_prime = ["NA" if i not in ["Y","N"] else i for i in a]
a_prime
```

```python
#use Map and lambda
a_prime2 = list(map(lambda i : "NA" if i not in ["Y","N"] else i ,a))
a_prime2
```

```python
# pandas apply 
func = lambda i : np.nan if i not in ["Y","N"] else i
db1['OWN_OCCUPIED'] = db1['OWN_OCCUPIED'].apply(func)
```
