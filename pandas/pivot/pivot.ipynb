```python
import pandas as pd 
import numpy as np 
```

```python
data = { 'names' : ['ali', 'reza' , 'rana' , 'ali', 'reza'] ,
        'sales' : [20,5,10,8,6] ,
        'amount' : [180,150,230,320,230]}
```

```python
df = pd.DataFrame(data)
df
```

```python
df.pivot_table(values = ['sales'] , index= ['names'] , aggfunc=np.sum)
```

```python
df.pivot_table(values = ['sales' , 'amount' ] , index= ['names'] , aggfunc={'sales':np.sum , 'amount':np.mean })
```
