```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.metrics import accuracy_score
iris = load_iris()
X = iris.data
y = iris.target
feature_names = iris.feature_names

```

### make sampling ----

```python
# Shuffle the data
np.random.seed(42)  # Set a seed for reproducibility
indices = np.random.permutation(len(X))
X_shuffled = X[indices]
y_shuffled = y[indices]
```

```python
# Create a Pandas DataFrame
iris_df = pd.DataFrame(data=X, columns=feature_names)
iris_df['target'] = y

# Visualize the data using Seaborn
plt.figure(figsize=(12, 8))

# Pairplot
sns.pairplot(iris_df, hue='target', palette="viridis")
plt.title("Iris Dataset - Pairplot")
plt.show()
```

```python
#sampling 
X_train, X_test, y_train, y_test = train_test_split(X_shuffled, y_shuffled, test_size=0.2, random_state=42)
```

```python
clf = LogisticRegression(max_iter = 1000 , verbose=1 ).fit(X_train, y_train)
```

```python
clf.predict([X_test[0]])
```

```python
y_test
```

### evaluate model with mse and r2

```python
pred_y_test = clf.predict(X_test)
accuracy = accuracy_score(y_test, pred_y_test)
accuracy
```

```python
mse = mean_squared_error(pred_y_test , y_test)
```

```python
mse
```

```python
r2 = r2_score(pred_y_test , y_test)
```
